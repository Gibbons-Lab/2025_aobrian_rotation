```python
import matplotlib.pyplot as plt
import pandas as pd


# read microbes hba1c regression
df = pd.read_csv("microbe_hits.csv")

# Sort microbes by beta values for better visualization
df_sorted = df.sort_values(by="beta", ascending=True)

# Plot beta values for each microbe
plt.figure(figsize=(12, 14))
plt.barh(df_sorted["independent_feature"], df_sorted["beta"], color="skyblue")
plt.xlabel("Beta Value (Effect on Statin)")
plt.ylabel("Microbes")
plt.title("Relationship Between Significant Microbes and Statin Users")
plt.axvline(x=0, color="gray", linestyle="--")  # Reference line at zero
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# read microbes hba1c regression
df = pd.read_csv("metabolite_hits.csv")

# Sort microbes by beta values for better visualization
df_sorted = df.sort_values(by="s_beta", ascending=True)

# Plot beta values for each microbe
plt.figure(figsize=(10, 12))
plt.barh(df_sorted["independent_feature"], df_sorted["s_beta"], color="skyblue")
plt.xlabel("Beta Value (Effect on Statin)")
plt.ylabel("Metabolites")
plt.title("Relationship Between Significant Metabolites and Statin Users")
plt.axvline(x=0, color="gray", linestyle="--")  # Reference line at zero
plt.grid(axis="x", linestyle="--", alpha=0.7)

# make a new graph that only plots beta whos absolute value is greater than 0.2
df_sorted = df_sorted[abs(df_sorted["s_beta"]) > 0.2]
plt.figure(figsize=(10, 13))
plt.barh(df_sorted["independent_feature"], df_sorted["s_beta"], color="skyblue")
plt.xlabel("Beta Value (Effect on Statin)", fontsize=18)
plt.ylabel("Metabolites", fontsize=18)
plt.title("Relationship Between Significant Metabolites and Statin Users")
plt.axvline(x=0, color="gray", linestyle="--")  # Reference line at zero
plt.grid(axis="x", linestyle="--", alpha=0.7)
# increase font size on y tick lables
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)

# Show the plot
plt.show()
```

```python
import seaborn as sns
# Read the data
df = pd.read_csv("hmgxa1c_microbes.csv")

# sort by beta values
df = df.sort_values(by="beta", ascending=True)

# Create a new DataFrame for plotting
plot_data = df[["dependent_feature", "beta", "interaction_beta"]]

# Set the dependent feature as the index and rename columns for clarity
plot_data = plot_data.set_index("dependent_feature")
plot_data.columns = ["Beta (HMG)", "Beta (Interaction A1c * HMG)"]

# Plot the heatmap
plt.figure(figsize=(6, 14))
sns.heatmap(plot_data, cmap="coolwarm", center=0, annot=True, fmt=".2f", linewidths=0.5)
plt.title("Relationship Between Significant Microbes and HMGXa1c")
plt.xlabel("Beta Values")
plt.ylabel("Dependent Features (Microbes)")
plt.show()
```
